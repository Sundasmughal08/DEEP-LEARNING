```python
df = df.fillna(df.median(numeric_only=True))
```

```python
df.info()
```

```python

# Features (X) and target (y)
X = df.drop("TenYearCHD", axis=1)   # target column remove
y = df["TenYearCHD"]                 # target column

# 80-20 split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,      # 20% testing
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
train_df = X_train.copy()
train_df["TenYearCHD"] = y_train
val_df = X_test.copy()
val_df["TenYearCHD"] = y_test
```

```python
def dataframe_to_dataset(df):
    df = df.copy()
    labels = df.pop('TenYearCHD')
    ds = tf.data.Dataset.from_tensor_slices(
        (dict(df), labels)
    )
    ds = ds.shuffle(buffer_size=len(df))
    return ds
```

```python
train_ds = dataframe_to_dataset(train_df)
val_ds   = dataframe_to_dataset(val_df)
train_ds = train_ds.batch(32)
val_ds   = val_ds.batch(32)
```

```python
from keras.utils import FeatureSpace

feature_space = FeatureSpace(
    features={
        # Categorical
        'male': 'integer_categorical',
        'currentSmoker': 'integer_categorical',
        'BPMeds': 'integer_categorical',
        'prevalentStroke': 'integer_categorical',
        'prevalentHyp': 'integer_categorical',
        'diabetes': 'integer_categorical',
        'education': 'integer_categorical',

        # Numerical
        'age': 'float_normalized',
        'cigsPerDay': 'float_normalized',
        'totChol': 'float_normalized',
        'sysBP': 'float_normalized',
        'diaBP': 'float_normalized',
        'BMI': 'float_normalized',
        'heartRate': 'float_normalized',
        'glucose': 'float_normalized',
    }
)


train_ds_with_no_labels = train_ds.map(lambda x, y: x)

feature_space.adapt(train_ds_with_no_labels)


for x, _ in train_ds.take(1):
    preprocessed_x = feature_space(x)
    print(preprocessed_x)

preprocessed_train_ds = train_ds.map(
    lambda x, y: (feature_space(x), y),
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)

preprocessed_val_ds = val_ds.map(
    lambda x, y: (feature_space(x), y),
    num_parallel_calls=tf.data.AUTOTUNE
).prefetch(tf.data.AUTOTUNE)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(
    preprocessed_train_ds,
    validation_data=preprocessed_val_ds,
    epochs=10
)
```

```python
training_model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

training_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

training_model.summary()

history = training_model.fit(
    preprocessed_train_ds,
    validation_data=preprocessed_val_ds,
    epochs=10
)
```

```python

for batch_x, _ in train_ds.take(1):
    example_batch = batch_x
    break

inputs = {
    key: tf.keras.Input(shape=(1,), name=key, dtype=value.dtype)
    for key, value in example_batch.items()
}

def preprocess_fn(inputs):
    return feature_space(inputs)
x = tf.keras.layers.Lambda(preprocess_fn)(inputs)
outputs = training_model(x)
inference_model = tf.keras.Model(inputs=inputs, outputs=outputs)
inference_model.summary()
```

```python
#Model training
training_model.fit(
    preprocessed_train_ds,
    validation_data=preprocessed_val_ds,
    epochs=10,
    verbose=2

)
```

```python
sample= {
    'male': 1,
    'age': 65.0,
    'education': 1,
    'currentSmoker': 1,
    'cigsPerDay': 25.0,
    'BPMeds': 1,
    'prevalentStroke': 1,
    'prevalentHyp': 1,
    'diabetes': 1,
    'totChol': 300.0,
    'sysBP': 180.0,
    'diaBP': 110.0,
    'BMI': 35.0,
    'heartRate': 95.0,
    'glucose': 180.0
}
```

```python
input_dict = {
    name: tf.convert_to_tensor([value])
    for name, value in sample.items()
}
```

```python
prediction = inference_model.predict(input_dict)

print(
    f"This patient has a {100 * prediction[0][0]:.2f}% probability "
    "of having heart disease."
)
```
